```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
import numpy as np

from skimage import io
from skimage import data
from scipy.signal import convolve2d
```

```python
# need to run twice to generate image

image = io.imread('dog.jpg', as_gray=True)

# 3x3 kernel with weights, [0,0,0;0,1,0;0,0,0] is the identity kernel, doesn't change anything except makes image 2 pixels smaller
# for this example, as i sweep across image, if i see a bright pixel followed by a dark pixel, the operation of this convolution is a really big number - essentially horizontal edge detection
kernel = [[0.0, 0.0, 0.0],
          [1.5, 0.0, -1.5],
          [0.0, 0.0, 0.0]]

features = convolve2d(image, kernel)
#Normalize between 0.0 and 1.0
features = (features - np.min(features)) / np.ptp(features)
io.imshow(features, cmap='gray')
```

```python
# need to run twice to generate image
# another kernel example

image = io.imread('dog.jpg', as_gray=True)

# 3x3 kernel with weights, [0,0,0;0,1,0;0,0,0] is the identity kernel, doesn't change anything except makes image 2 pixels smaller
# for this example, as i sweep across image, if i see a bright pixel followed by a dark pixel, the operation of this convolution is a really big number - essentially horizontal edge detection
# the weights are the kernel, and the network will learn what convolution to perform
# need a convolution if you want to understand relationships in 2D space, since perceptrons flatten the data into an array

kernel = [[0.5, 0.0, 0.5],
          [-1.5, 0.0, -1.5],
          [0.5, 0.0, 0.5]]

features = convolve2d(image, kernel)
#Normalize between 0.0 and 1.0
features = (features - np.min(features)) / np.ptp(features)
io.imshow(features, cmap='gray')
```
